# Cria arquivo com informações iniciais dos professores

````python
%%time
import grequests
from lxml import html
import pandas as pd

URL = "https://sigaa.ufersa.edu.br/sigaa/public/docente/busca_docentes.jsf"
URL_BASE = "https://sigaa.ufersa.edu.br"

def get_session():
    '''
    '''
    session = grequests.Session()
    return session
        

def get_view_state(session):
    '''
    '''
    response = session.get(URL) 
    body = html.fromstring(response.text)
    viewState = body.xpath('//input[@id="javax.faces.ViewState"]/@value')[0]
    return viewState

def get_ufersa_departament_page():
    '''
    '''
    session = get_session() #it is necessary for jsession on server.
    viewState = get_view_state(session)
    rs = set()
    
    for k,v in DEPARTAMENTS.items():
        params = {'form': 'form', 
                  'form:nome':'',
                  'form:departamento':k, 
                  'form:buscar':'Buscar',
                  'javax.faces.ViewState':viewState}
        
        rs.add(session.post(URL, data=params))
    return rs

def create_ufersa_professors_csv(list_professors, labels):
    '''
    '''
    csv_file = pd.DataFrame.from_records(list_professors, columns=labels)
    csv_file.to_csv('ufersa_teachers.csv', sep=',', encoding='utf-8', index=False)
    print "Done!"    

def get_siape_professors(link_paginas):
    '''
    '''
    SIAPE = 1
    siapes = [int(link.split('=')[SIAPE]) for link in link_paginas]
    return siapes
    
def insert_URL_BASE(url_base, incomplete_links):
    '''
    '''
    complet_links = [url_base+i for i in incomplete_links]
    return complet_links
    
def get_list_ufersa_professors():
    '''
    '''
    nomes = []
    departamentos = []
    fotos = []
    paginas = []
    
    responses = get_ufersa_departament_page()
    for rs in responses:
        departament_page = html.fromstring(rs.text)
        nomes.extend(departament_page.xpath('//span[@class="nome"]/text()'))
        departamentos.extend(departament_page.xpath('//span[@class="departamento"]/text()'))
        fotos.extend(departament_page.xpath('//td[@class="foto"]/img/@src'))
        paginas.extend(departament_page.xpath('//span[@class="pagina"]/a/@href'))
    
    siapes = get_siape_professors(paginas)
    paginas = insert_URL_BASE(URL_BASE, paginas)
    fotos = insert_URL_BASE(URL_BASE, fotos)
    list_professors = zip(siapes, nomes, departamentos, fotos, paginas)
    columns = ['siapes','nomes','departamentos','fotos','paginas']
    create_ufersa_professors_csv(list_professors, columns)

get_list_ufersa_professors()
````
